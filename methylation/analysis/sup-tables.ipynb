## SC

```r
samp_levels <- c("d5_3a", "d5_3b", "d5_wt", "e7.5_ecto", "e7.5_meso", "e7.5_endo")
```

```r
sc_stats <- db_f@cells %>% 
    filter(day %in% c("d5", "e7.5"))
```

```r
invivo_sort <- fread(here("data/cells_germ_layer_invivo.tsv")) %>% as_tibble()
```

```r
sc_stats <- sc_stats %>% 
    left_join(invivo_sort) %>% 
    replace_na(replace = list(germ_layer = "other")) %>%
    mutate(sort = ifelse(day == "e7.5", germ_layer, sort)) 
#     filter(!(experiment %in% paste0("experiment", c(5,6,8))))
```

```r
new_plates <- sc_stats %>% 
    arrange(plate, as.numeric(cell_num)) %>% 
    distinct(plate) %>% 
    mutate(new_plate = paste0("plate_", str_pad(1:n(), width=2, pad = 0)))
new_plates
```

```r
sc_stats1 <- sc_stats %>% 
    left_join(new_plates) %>% 
    select(-plate) %>% 
    rename(plate = new_plate) %>% 
    mutate(cell_id = paste0(plate, ".", cell_num)) %>% 
    arrange(plate, as.numeric(cell_num)) 
sc_stats1 %>% select(plate, cell_id, cell_num) %>% head()
```

```r
sc_sup_tab <- sc_stats1 %>% 
    arrange(plate, as.numeric(cell_num)) %>%
    mutate(sort = gsub("blood\\?", "blood", sort)) %>%
    mutate(
        line = gsub("ko3a", "3a-/-", line), 
        line = gsub("ko3b", "3b-/-", line) 
            ) %>%
    select(
        cell_id, 
        day, 
        line, 
        sort,
        EPCAM = APC_A,
        CXCR4 = BV421_A,
        plate, 
        experiment,
        amp_batch=batch_id, 
        index1 = index1.seq, 
        index2 = index2.seq, 
        total_reads, 
        mapped_reads, 
        uniq_frac, 
        cpg_num, 
        CHH, 
        CHG, 
        CpG) 

sc_sup_tab %>% head
```

## Bulk

```r
used_tracks <- tracks_key %>% filter(day %in% c("d0S", paste0("d", 0:6)), !grepl("N15", line)) %>% filter(line != "ko1")
used_tracks <- used_tracks  %>% 
    select(track = track_name, day, line, sort=sort_fraction, capture) %>%     
    mutate(sort = ifelse(sort == "", "all", sort)) %>%
    arrange(day, line, sort) %>%
    group_by(day, line) %>% 
    mutate(samp_id = paste0(day, "_", line, ".", 1:n())) %>% 
    ungroup() %>%
    mutate(
        line = gsub("ko3a", "3a-/-", line), 
        line = gsub("ko3b", "3b-/-", line), 
        line = gsub("tko", "TKO", line)
    ) 
used_tracks %>% head()
```

```r
bulk_sup_tab <- used_tracks %>% left_join(
    track_stats %>% 
        select(
            track, 
            total_reads, 
            mapped_reads, 
            uniq_frac,
            cpg_num = cg_num, 
            meth_calls, 
            CHH, 
            CHG,
            CpG = global_avg_meth
        )
    ) %>% 
    select(-track) %>% 
    select(samp_id, everything())
bulk_sup_tab
```

```r
# writexl::write_xlsx(list("Bulk" = bulk_sup_tab, "Single cell" = sc_sup_tab), path = here("output/S2 - Methylation Samples.xlsx"))
```

```r
plates_table <- sc_sup_tab %>% 
    mutate(sort = ifelse(grepl("^CXCR", sort), sort, "index")) %>% 
    group_by(plate, day, line, sort) %>% 
    summarise(
        n_cells = n(), 
        capture = "wgbs", 
        total_reads = mean(total_reads), 
        mapped_reads = mean(mapped_reads), 
        uniq_frac = mean(uniq_frac), 
        meth_calls = mean(cpg_num), 
        cpg_num = mean(cpg_num), 
        CHH = mean(CHH), 
        CHG = mean(CHG, na.rm=TRUE), 
        CpG = mean(CpG), 
        .groups="drop") %>% 
    arrange(day, line, sort) %>% 
    group_by(day, line) %>% 
    mutate(samp_id = paste0(day, "_", line, ".sc", 1:n())) %>% 
    ungroup() %>% 
    select(-plate)
plates_table
```

```r
comb_table <- bind_rows(plates_table %>% mutate(type = "sc"), bulk_sup_tab %>% mutate(type = "bulk")) %>% arrange(samp_id) %>% select(samp_id, day, line, sort, type, n_cells, everything())
comb_table
```

```r
writexl::write_xlsx(comb_table, path = here("output/S2 - Methylation Samples.xlsx"))
```
